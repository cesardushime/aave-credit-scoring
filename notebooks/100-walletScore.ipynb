```python
# Read transactions json file into a DataFrame
import json

path = 'C:\\Users\\Cesar Dushimimana\\Documents\\aave-credit-scoring\\data\\transactions.json'

```

```python
def load_json_data(path):
    with open(path, "r") as f:
        raw_data = json.load(f)
    return pd.DataFrame(raw_data)

df = load_json_data(path)
print(f"✅ Loaded {len(df)} rows")
df.head()
```

```python
count = 0
for ids in sample_wallets:
    for user_wallet in df['userWallet']:
        if ids == user_wallet:
            count += 1

print(f"✅ Number of matching wallet IDs: {count}")


print(f"Total in sample_wallets: {len(sample_wallets)}")
print(f"Matching IDs: {count} in {len(df)} transactions")
```
